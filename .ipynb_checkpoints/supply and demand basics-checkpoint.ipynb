# Overview

The most fundamental basic analysis of economics is supply and demand. 

```python
# load plotting tools
from matplotlib import pyplot as plt
import numpy as np
```

Demand

```python
# demand function
def demand(x):
    return -x + 10

# plot demand function
plt.xkcd()

data = np.arange(1, 10, 1)

dem = plt.plot(data, demand(data))

plt.axis((0,10,0,10))

plt.xticks([])
plt.yticks([])

plt.setp(dem, color='dodgerblue', linewidth=5.0)

plt.xlabel("Quantity")
plt.ylabel("Price")

plt.annotate('Demand', xy=(1.2,8.5))

plt.show()
```

```python
# demand function
def d1(x):
    return -x + 10

def d2(x):
    return -x + 5

# plot demand function
plt.xkcd()

d1_data = np.arange(5, 10, 1)
d2_data = np.arange(1, 6, 1)

plt.plot(d1_data, d1(d1_data), "dodgerblue", d2_data, d2(d2_data), "red")

plt.axis((0,10,-2,7))

plt.xticks([])
plt.yticks([])

plt.xlabel("Quantity")
plt.ylabel("Price")

plt.annotate("D1", xy=(5.1,5))
plt.annotate("D2", xy=(1.2,3.9))
plt.annotate("Decrease\nin Demand", xy=(2.9,2.3), arrowprops=dict(arrowstyle='->'), xytext=(5,2))

plt.show()
```

```python
# supply function
def supply(x):
    return x

# plot demand function
plt.xkcd()

supp = plt.plot(data, supply(data))

plt.axis((0,10,0,10))

plt.xticks([])
plt.yticks([])

plt.setp(supp, color='limegreen', linewidth=5.0)

plt.xlabel("Quantity")
plt.ylabel("Price")

plt.annotate('Supply', xy=(1.8,1.5))

plt.show()
```

```python
# demand function
def d1(x):
    return -x + 10

def d2(x):
    return -x + 5

# plot demand function
plt.xkcd()

d1_data = np.arange(5, 10, 1)
d2_data = np.arange(1, 6, 1)

plt.plot(d1_data, d1(d1_data), "dodgerblue", d2_data, d2(d2_data), "red")

plt.axis((0,10,-2,7))

plt.xticks([])
plt.yticks([])

plt.xlabel("Quantity")
plt.ylabel("Price")

plt.annotate("D1", xy=(5.1,5))
plt.annotate("D2", xy=(1.2,3.9))
plt.annotate("Increase\nin Supply", xy=(2.9,2.3), arrowprops=dict(arrowstyle='->'), xytext=(5,2))

plt.show()
```

```python

# plot demand function
plt.xkcd()

plt.plot(data, demand(data), 'dodgerblue', data, supply(data), 'limegreen', linewidth=5.0)

plt.axis((0,10,0,10))

plt.xticks([])
plt.yticks([])

plt.xlabel("Quantity")
plt.ylabel("Price")

plt.annotate('Equilibrium', xy=(5,5))
plt.annotate('Demand', xy=(1.5,8.5))
plt.annotate('Supply', xy=(1.8,1.5))

plt.show()
```
